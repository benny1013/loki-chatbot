### 中正心理系小助手
1. 背景與動機：
- 有鑑於本系系網資訊過於繁雜且本學期即將結束，邁入下個學期，有許多大學部及研究所學生，又或者是已經確定錄取本系研究所的學生，將會頻繁透過系網了解教授相關資訊，因此思考將系網的活動與教授資訊進行整合，加上系上分工較細，系辦不見得全盤了解所有學生事務，故也將經常使用到的相關聯絡資訊融入其中。
- 此外，現在正逢下學期選課期間，使用單一入口登入選課系統反覆查看選課資訊相當不便利，且上學期有開設的課程需透過尋找各系所代碼的方式才能正確取的過往的開課資訊，因此，我也將大學部與研究所分別在上學期與下學期有開設的課程納入其中。
- 綜合以上，此聊天機器人能提供使用者每位教授的相關資訊、系上近期演講活動、系上各承辦單位的聯絡資訊以及113學年度心理系及心理所開設之課程。
2. 目的與功能：
- 不管是在下個學期或下個學年都會有新生加入心理系的大家庭，根據我過往的經驗，要查找自己感興趣的教授時，需耗費大量時間進入系網點選，而此聊天機器人能提供使用者查詢自身感興趣的領域有哪些教授，在進一步了解該領域內的教授各自的研究專長為何，若想加入該教授實驗室也可透過詢問電子信箱的方式獲得正確訊息。
- 再者，系上通常會以信件通知領取文件等相關訊息，且每間教室之網路密碼也需透過季送信件至指定的電子郵件才能獲得，對於心理系學生來說，最常會詢問的工作單位還有獎學金小組，故此聊天機器人亦能提供使用者系辦、電腦助教、獎學金小組之信箱。
- 最後，在學期間系上經常會舉辦大大小小的演講活動，有時會透過信箱發送告知有興趣的同學，但有時只公告在系網，為了減少系網如此多資訊混雜，導致難以找到的狀況，我利用爬蟲將系網上公告近期的演講與活動放入聊天機器人的意圖中，一來，能增加同學取得資訊的便利性，二來，隨著時間推移，聊天機器人的回覆能進行更新，讓每位學生知道系上所有的活動，若想參與，也能盡早安排時間。
